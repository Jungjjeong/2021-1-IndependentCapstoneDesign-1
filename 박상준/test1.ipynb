```python
import networkx as nx
from IPython.core.display import Image
from networkx.drawing.nx_pydot import to_pydot
import graphviz
import numpy as np

g3 = nx.Graph()

file = open('2.txt', "r")


table_data =[]  

for line in file:
    tmp_array = line.strip().split("\t")
    table_data.append(tmp_array)


arr = np.array(table_data)
'''
x = str(arr[0:1,1])
y = str(arr[0:1,2])
z = str(arr[0:1,3])

g3.add_node(x)
g3.add_node(y)
g3.add_node(z)
g3.add_edge(x,y)
g3.add_edge(y,z)
d3 = to_pydot(g3)
d3.set_dpi(600)
d3.set_rankdir("LR")
Image(d3.create_png(), width=800)
'''


for i in range(0,50):
    x = str(arr[i:i+1,1])
    y = str(arr[i:i+1,2])
    z = str(arr[i:i+1,3])
    g3.add_node(x)
    g3.add_node(y)
    g3.add_node(z)
    g3.add_edge(x,y)
    g3.add_edge(y,z)
d3 = to_pydot(g3)
d3.set_dpi(600)
d3.set_rankdir("LR")
Image(d3.create_png(), width=800)
```
